```python
#read P  L March 2021.csv

import pandas as pd

p_l_march_2021_df = pd.read_csv("P  L March 2021.csv", header=0)
```

```python
# check the percentage of missing values in each column
p_l_march_2021_df.isnull().mean() * 100
```

```python
#drop Style Id column
p_l_march_2021_df.drop(columns=["Style Id"], inplace=True)
```

```python
#convert weight to float
#if nil, replace with 0.0. use regex to replace nil with 0.0
p_l_march_2021_df["Weight"] = p_l_march_2021_df["Weight"].replace(
    "Nill", "0.0").astype(float)

#convert TP 1 to int
# replace #VALUE! with 0 and remove the decimal part of the number
p_l_march_2021_df["TP 1"] = p_l_march_2021_df["TP 1"].replace(
    "#VALUE!", "0").astype(float).astype(int)

#convert TP 2 to int
p_l_march_2021_df["TP 2"] = p_l_march_2021_df["TP 2"].replace("#VALUE!", "0").astype(float).astype(int)

#convert MRP Old to int
p_l_march_2021_df["MRP Old"] = p_l_march_2021_df["MRP Old"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Final MRP Old to int
p_l_march_2021_df["Final MRP Old"] = p_l_march_2021_df["Final MRP Old"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Ajio MRP to int
p_l_march_2021_df["Ajio MRP"] = p_l_march_2021_df["Ajio MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Amazon MRP to int
p_l_march_2021_df["Amazon MRP"] = p_l_march_2021_df["Amazon MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Amazon FBA MRP to int
p_l_march_2021_df["Amazon FBA MRP"] = p_l_march_2021_df["Amazon FBA MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Flipkart MRP to int
p_l_march_2021_df["Flipkart MRP"] = p_l_march_2021_df["Flipkart MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Limeroad MRP to int
p_l_march_2021_df["Limeroad MRP"] = p_l_march_2021_df["Limeroad MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Myntra MRP to int
p_l_march_2021_df["Myntra MRP"] = p_l_march_2021_df["Myntra MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Paytm MRP to int
p_l_march_2021_df["Paytm MRP"] = p_l_march_2021_df["Paytm MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
#Snapdeal MRP to int
p_l_march_2021_df["Snapdeal MRP"] = p_l_march_2021_df["Snapdeal MRP"].replace(
    "Nill", "0.0").astype(float).astype(int)
```

```python
# load dataframe to mysql database
from sqlalchemy import create_engine

# create table amazon_sales and load data into it
engine = create_engine("mysql+pymysql://root@localhost:3306/e-commerce_sales")

p_l_march_2021_df.to_sql(name="p_l_march_2021", con=engine, if_exists="replace", index=False)

engine.dispose()
```
