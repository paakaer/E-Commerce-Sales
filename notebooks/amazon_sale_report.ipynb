```python
import pandas as pd

amazon_sale_report_df = pd.read_csv("Amazon Sale Report.csv")
```

```python
#check the percentage of missing values in each column
amazon_sale_report_df.isnull().mean() * 100
```

```python
amazon_sale_report_df["Courier Status"].unique()
```

```python
# fix "Courier Status" column nan -> Unknown
amazon_sale_report_df["Courier Status"] = amazon_sale_report_df["Courier Status"].fillna("Unknown")
```

```python
# fix currency column
# print rows with currency  == nan
amazon_sale_report_df[amazon_sale_report_df["currency"].isna()]
```

```python
# fix currency column nan -> INR
amazon_sale_report_df["currency"] = amazon_sale_report_df["currency"].fillna("INR")
# fix Amount column nan -> 0.0
amazon_sale_report_df["Amount"] = amazon_sale_report_df["Amount"].fillna(0.0)
```

```python
# fix Amount column to float
# every value should follow this format -> .2f
amazon_sale_report_df["Amount"] = amazon_sale_report_df["Amount"].astype(float)

# remove leading and trailing whitespace
amazon_sale_report_df["Date"].str.strip()
# convert to datetime format
amazon_sale_report_df["Date"] = pd.to_datetime(amazon_sale_report_df["Date"], format="%m-%d-%y")
```

```python
#drop the following columns: promotion-ids, fulfilled-by, Unnamed: 22, ship-service-level, Sales Channel, ship-city, ship-state, ship-postal-code, ship-country, Status
amazon_sale_report_df.drop(columns=["promotion-ids", "fulfilled-by", "Unnamed: 22", "ship-service-level", "Sales Channel ", "ship-city", "ship-state", "ship-postal-code", "ship-country", "Status"], inplace=True)
#check the percentage of missing values in each column again
amazon_sale_report_df.isnull().mean() * 100
```

```python
#drop index column and rename currency column to Currency
amazon_sale_report_df.drop(columns=["index"], inplace=True)
amazon_sale_report_df.rename(columns={"currency": "Currency"}, inplace=True)
```

```python
#load dataframe to mysql database
from sqlalchemy import create_engine

#create table amazon_sales and load data into it
engine = create_engine("mysql+pymysql://root@localhost:3306/e-commerce_sales")
```

```python
amazon_sale_report_df.to_sql(
    "amazon_sales", con=engine, if_exists="replace", index=False)
```
