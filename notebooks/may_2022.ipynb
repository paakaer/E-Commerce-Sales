```python
#read May-2022.csv

import pandas as pd

may_2022_df = pd.read_csv("May-2022.csv", header=0)
```

```python
#check the percentage of missing values in each column
may_2022_df.isnull().mean() * 100
```

```python
#drop duplicates
may_2022_df.drop_duplicates(inplace=True)
```

```python
#remove the noise. i.e unnecessary columns like:
# Style Id

may_2022_df.drop(columns=["Style Id"], inplace=True)
```

```python
may_2022_df
```

```python
# load dataframe to mysql database
from sqlalchemy import create_engine

# create table amazon_sales and load data into it
engine = create_engine("mysql+pymysql://root@localhost:3306/e-commerce_sales")

may_2022_df.to_sql(name="may_2022", con=engine, if_exists="replace", index=False)

engine.dispose()
```
