```python
#read Sale Report.csv file
import pandas as pd

sale_report_df = pd.read_csv("Sale Report.csv", header=0)
```

```python
#check the percentage of missing values in each column
sale_report_df.isnull().mean() * 100
```

```python
#drop rows that have missing values in the SKU Code column
sale_report_df.dropna(subset=["SKU Code"], inplace=True)
#check the percentage of missing values in each column again
sale_report_df.isnull().mean() * 100
```

```python
# cast columns to appropriate data types
sale_report_df["SKU Code"].astype(str)
sale_report_df["Design No."].astype(str)
sale_report_df["Stock"].astype(int)
sale_report_df["Category"].astype(str)
sale_report_df["Size"].astype(str)
sale_report_df["Color"].astype(str)
```

```python
sale_report_df
```

```python
#reset the index of the DataFrame
sale_report_df.reset_index(drop=True, inplace=True)
```

```python
# load dataframe to mysql database
from sqlalchemy import create_engine

# create table amazon_sales and load data into it
engine = create_engine("mysql+pymysql://root@localhost:3306/e-commerce_sales")

sale_report_df.to_sql(name="sale_report", con=engine, if_exists="replace", index=False)

engine.dispose()
```
